23) Crear un diccionario utilizando la lista crada en el punto 1, asignandole la clave "ciudad". Agregar tambien otras claves, como puede ser "Pais" y "Continente".

```python
listaContinentes = ["America", "Europa", "Asia"]
listaCiudades = ["Roma", "Buenos Aires", "Montevideo", "Madrid912"]

largo = max(len(listaPaises), len(listaCiudades), len(listaContinentes))

listaDicc = []

for i in range(largo):
    dicc = {
        "Pais": listaPaises[i] if i < len(listaPaises) else "No insertado",
        "ciudad": listaCiudades[i] if i < len(listaCiudades) else "No insertado",
        "Continente": listaContinentes[i] if i < len(listaContinentes) else "No insertado"
    }
    listaDicc.append(dicc)

print(listaDicc)
```

24) Imprimir las claves del diccionario

```python
print(dicc.keys())
```

25) Imprimir las ciudades a través de su clave

```python
for d in listaDicc:
    if(d["ciudad"] == "No insertado"):
        break
    print(d["ciudad"])
```

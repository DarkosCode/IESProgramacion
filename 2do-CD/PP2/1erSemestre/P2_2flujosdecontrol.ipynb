11) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
n = 99
while(n <= 300):
    n += 1
    if(n % 12 != 0):
        continue
    print(f"el {n} es divisible por 12!")
    
```
